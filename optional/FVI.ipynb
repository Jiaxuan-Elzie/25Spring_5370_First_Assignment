```python
"""
Fitted Value Iteration (FVI) — Constrained Portfolio Optimization
Assignment 1 | Discrete-time asset allocation, Section 8.4 (Rao & Jelvis)

Assignment specification:
  - Assets   : k=0 (cash, rate r)  +  k=1…n (risky),   n > 2,  n < 5
  - Returns  : risky asset k  ~  N(a[k], s[k]),   s[k] = VARIANCE (per period)
  - Weights  : p[0]+p[1]+…+p[n] = 1,  p[k] ≥ 0  (no short-selling)
  - Constraint: |Δp[k]| ≤ 10% per period
  - Utility  : CARA,  U(W) = −exp(−αW)/α   (absolute risk averse)
  - Horizon  : T < 10
"""

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from typing import Tuple, List, Dict, Optional, Callable


# ════════════════════════════════════════════════════════════════════
#  Value function approximator
# ════════════════════════════════════════════════════════════════════

class ValueNet(nn.Module):
    """Fully-connected network approximating V_t(W, p)."""

    def __init__(self, input_dim: int, hidden_dims: List[int] = [128, 128, 64]):
        super().__init__()
        layers, prev = [], input_dim
        for h in hidden_dims:
            layers += [nn.Linear(prev, h), nn.Tanh()]
            prev = h
        layers.append(nn.Linear(prev, 1))
        self.net = nn.Sequential(*layers)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.net(x).squeeze(-1)


# ════════════════════════════════════════════════════════════════════
#  FVI solver
# ════════════════════════════════════════════════════════════════════

class FVI:
    """
    Fitted Value Iteration for constrained multi-asset portfolio optimization.

    State  : (W_t, p_t)   wealth × portfolio weights  (n+1 assets incl. cash)
    Action : Δp_t         |Δp[k]| ≤ delta_max,  Σ Δp[k] = 0
    Reward : terminal CARA utility  U(W_T)

    Why these parameters work  (signal-to-noise analysis):
      Per-period signal  ≈ exp(−αW)·(a[k]−r)  ≈ 0.011
      MC noise (SE)      ≈ α·W·σ·exp(−αW)/√n_mc ≈ 0.003  (antithetic, n_mc=300)
      SNR ≈ 4  →  SLSQP can reliably find the optimum
      (Monthly dt=1/12 would give SNR≈0.25 → optimization fails)
    """

    def __init__(
        self,
        n_risky:   int   = 3,       # risky assets  [assignment: n > 2,  n < 5]
        T:         int   = 5,       # horizon        [assignment: T < 10]
        alpha:     float = 2.0,     # CARA risk-aversion coefficient
        delta_max: float = 0.10,    # max per-asset adjustment per period
        r:         float = 0.03,    # risk-free rate per period (cash k=0)
        a:  Optional[np.ndarray] = None,    # mean returns   shape (n_risky,)
        s:  Optional[np.ndarray] = None,    # VARIANCES       shape (n_risky,)
        corr: Optional[np.ndarray] = None,  # correlation matrix (default: identity)
        seed: int = 42,
    ):
        # ── Assignment constraint checks ──────────────────────────
        assert 2 < n_risky < 5, \
            f"Assignment requires 2 < n < 5 risky assets; got n_risky={n_risky}"
        assert T < 10, \
            f"Assignment requires T < 10; got T={T}"

        np.random.seed(seed)
        torch.manual_seed(seed)

        self.n_risky   = n_risky
        self.n_assets  = n_risky + 1   # total slots including cash (k=0)
        self.T         = T
        self.alpha     = alpha
        self.delta_max = delta_max
        self.r         = r

        # ── Default market parameters (per-period, no dt scaling) ─
        #   These represent annual returns if 1 period = 1 year.
        #   s[k] = VARIANCE  (not std) per assignment specification.
        _a_def = np.array([0.08, 0.12, 0.10, 0.09])
        _s_def = np.array([0.04, 0.09, 0.06, 0.05])   # variances; std≈[.20,.30,.24,.22]

        self.a       = np.asarray(a if a is not None else _a_def[:n_risky], dtype=float)
        self.s       = np.asarray(s if s is not None else _s_def[:n_risky], dtype=float)
        self.std_dev = np.sqrt(self.s)   # std for sampling

        # Build covariance matrix from (std, corr)
        if corr is None:
            corr = np.eye(n_risky)
            if n_risky >= 2: corr[0, 1] = corr[1, 0] = 0.30
            if n_risky >= 3: corr[0, 2] = corr[2, 0] = 0.20; corr[1, 2] = corr[2, 1] = 0.35
        self.corr = np.asarray(corr, dtype=float)
        self.cov  = np.outer(self.std_dev, self.std_dev) * self.corr

        # ── FVI training hyper-parameters ─────────────────────────
        self.n_samples  = 500     # number of (W, p) states per backward step
        self.n_mc       = 300     # MC paths per Bellman eval (antithetic → ×2 effective)
        self.n_epochs   = 400     # NN epochs per backward step
        self.batch_size = 64
        self.lr         = 1e-3
        self.device     = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        print(f"Using device: {self.device}")

        self.value_nets: Dict[int, Tuple[nn.Module, float, float]] = {}
        self._print_summary()

    # ─────────────────────────────────────────────────────────────
    def _print_summary(self):
        print("═" * 62)
        print("  FVI Portfolio Optimizer  (Assignment 1, §8.4)")
        print(f"  Risky assets  : {self.n_risky}    T = {self.T}    α = {self.alpha}")
        print(f"  Cash rate r   : {self.r:.2%} per period")
        print(f"  Mean returns  : {np.round(self.a, 4)}")
        print(f"  Variances s   : {np.round(self.s, 4)}")
        print(f"  Std devs √s   : {np.round(self.std_dev, 4)}")
        print(f"  Δ_max         : ±{self.delta_max:.0%} per asset per period")
        print("═" * 62)

    # ─────────────────────────────────────────────────────────────
    # Sampling helpers
    # ─────────────────────────────────────────────────────────────

    def _sample_returns(self, n: int) -> np.ndarray:
        """Draw n i.i.d. return vectors ~ N(a, Cov).  Shape (n, n_risky)."""
        return np.random.multivariate_normal(self.a, self.cov, size=n)

    def _sample_antithetic(self, n: int) -> np.ndarray:
        """
        Antithetic variates: pairs (z, 2·a − z), both ~ N(a, Cov).
        Negatively correlated → reduces MC variance by ~50%.
        Returns shape (2·⌈n/2⌉, n_risky).
        """
        half = max((n + 1) // 2, 1)
        base = np.random.multivariate_normal(self.a, self.cov, size=half)
        anti = 2.0 * self.a - base          # mirror about mean
        return np.vstack([base, anti])

    # ─────────────────────────────────────────────────────────────
    # State transition  (Section 8.4 dynamics)
    # ─────────────────────────────────────────────────────────────

    def _full_ret(self, risky_ret: np.ndarray) -> np.ndarray:
        """Append cash return: [r, r_1, …, r_n]."""
        return np.concatenate([[self.r], risky_ret])

    def next_state(
        self,
        W: float,
        p: np.ndarray,          # (n_assets,) current weights
        delta_p: np.ndarray,    # (n_assets,) action
        risky_ret: np.ndarray,  # (n_risky,)  realized returns
    ) -> Tuple[Optional[float], Optional[np.ndarray]]:
        """
        Section 8.4 state transition:
            p_new = p + Δp                           rebalance
            R     = p_new · ret_full                 portfolio return
            W'    = W · (1 + R)                      wealth update
            p'_k  = p_new_k · (1 + ret_k) / (1 + R) weight drift
        """
        p_new = p + delta_p
        ret   = self._full_ret(risky_ret)
        R     = float(p_new @ ret)
        W_new = W * (1.0 + R)
        if W_new <= 1e-12:
            return None, None
        p_drift = p_new * (1.0 + ret) / (1.0 + R)
        return W_new, p_drift

    # ─────────────────────────────────────────────────────────────
    # Utility, features, value prediction
    # ─────────────────────────────────────────────────────────────

    def utility(self, W: float) -> float:
        """CARA: U(W) = −exp(−αW)/α   (absolute risk averse per assignment)."""
        return float(-np.exp(-self.alpha * W) / self.alpha)

    def _features(self, W: float, p: np.ndarray) -> np.ndarray:
        """State feature vector: [log W, p_0, p_1, …, p_n].  Shape (n_assets+1,)."""
        return np.concatenate([[np.log(max(W, 1e-12))], p], dtype=np.float32)

    def _predict(self, t: int, W: float, p: np.ndarray) -> float:
        """Evaluate V_t(W, p) via trained network (with de-normalisation)."""
        net, mu, sig = self.value_nets[t]
        x = torch.FloatTensor(self._features(W, p)).unsqueeze(0).to(self.device)
        with torch.no_grad():
            return float(net(x).item() * sig + mu)

    # ─────────────────────────────────────────────────────────────
    # Bellman expectation (Monte Carlo)
    # ─────────────────────────────────────────────────────────────

    def _bellman_mc(
        self,
        W: float,
        p: np.ndarray,
        delta_p: np.ndarray,
        t: int,
        rets: np.ndarray,       # pre-sampled, (n_mc, n_risky)
    ) -> float:
        """
        E[ V_{t+1}(W', p') | W, p, Δp ]  via MC.
        Samples are fixed for the entire SLSQP call → smooth landscape.
        """
        total = 0.0
        for ret in rets:
            W_new, p_new = self.next_state(W, p, delta_p, ret)
            if W_new is None:
                total += -1e4 / self.alpha          # ruin penalty
            elif t == self.T - 1:
                total += self.utility(W_new)         # terminal condition
            else:
                total += self._predict(t + 1, W_new, p_new)
        return total / len(rets)

    # ─────────────────────────────────────────────────────────────
    # Inner optimisation: best Δp for given (W, p, t)
    # ─────────────────────────────────────────────────────────────

    def optimize_delta_p(
        self,
        W: float,
        p: np.ndarray,
        t: int,
        n_mc: Optional[int] = None,
    ) -> Tuple[np.ndarray, float]:
        """
        Solve the constrained Bellman inner problem:
            max_{Δp}  E[ V_{t+1}(W', p') ]
            s.t.   Σ_k Δp_k = 0             (Σ weights preserved at 1)
                   |Δp_k|   ≤ delta_max      (10% per-asset turnover limit)
                   p_k + Δp_k ≥ 0            (no short-selling)

        Solver: SLSQP + antithetic MC sampling + 2 initial points.
        """
        n_mc = n_mc or self.n_mc
        rets = self._sample_antithetic(n_mc)     # fixed for this call

        def neg_obj(dp: np.ndarray) -> float:
            return -self._bellman_mc(W, p, dp, t, rets)

        bounds = [
            (max(-self.delta_max, -p[k]),
             min( self.delta_max,  1.0 - p[k]))
            for k in range(self.n_assets)
        ]
        cons = [{'type': 'eq', 'fun': lambda dp: float(dp.sum())}]

        best_val, best_dp = np.inf, np.zeros(self.n_assets)
        for x0 in self._init_starts(p):
            res = minimize(neg_obj, x0=x0, method='SLSQP',
                           bounds=bounds, constraints=cons,
                           options={'ftol': 1e-7, 'maxiter': 150, 'disp': False})
            if res.success and res.fun < best_val:
                best_val, best_dp = res.fun, res.x.copy()

        if np.isinf(best_val):          # all starts failed → hold
            best_dp  = np.zeros(self.n_assets)
            best_val = -neg_obj(best_dp)
        else:
            best_val = -best_val

        return best_dp, best_val

    def _init_starts(self, p: np.ndarray) -> List[np.ndarray]:
        """
        Two SLSQP starting points:
          (1) zero (hold current portfolio)
          (2) tilt: reduce lowest-return asset, increase highest-return asset
        """
        starts = [np.zeros(self.n_assets)]

        full_a = np.concatenate([[self.r], self.a])
        k_hi   = int(np.argmax(full_a))
        k_lo   = int(np.argmin(full_a))
        tilt   = np.zeros(self.n_assets)
        tilt[k_hi] =  min(self.delta_max, 1.0 - p[k_hi])
        tilt[k_lo] = -min(self.delta_max, p[k_lo])
        tilt -= tilt.mean()                         # enforce Σ = 0
        starts.append(tilt)

        return starts

    # ─────────────────────────────────────────────────────────────
    # FVI main loop  (backward induction)
    # ─────────────────────────────────────────────────────────────

    def train(
        self,
        n_samples:  Optional[int]   = None,
        n_mc:       Optional[int]   = None,
        n_epochs:   Optional[int]   = None,
        batch_size: Optional[int]   = None,
        lr:         Optional[float] = None,
    ) -> None:
        """
        Fitted Value Iteration — backward induction t = T−1, T−2, …, 0.

        Each step:
          1. Sample random (W_i, p_i) states
          2. For each state: solve max_{Δp} E[V_{t+1}]  (SLSQP + MC)
          3. Fit  ValueNet_t : state → Bellman target  (supervised regression)
        """
        if n_samples  is not None: self.n_samples  = n_samples
        if n_mc       is not None: self.n_mc       = n_mc
        if n_epochs   is not None: self.n_epochs   = n_epochs
        if batch_size is not None: self.batch_size = batch_size
        if lr         is not None: self.lr         = lr

        input_dim = 1 + self.n_assets   # log(W) + n_assets weights

        print(f"\n{'═'*62}")
        print(f"  Training FVI | n_samples={self.n_samples}  "
              f"n_mc={self.n_mc}  n_epochs={self.n_epochs}")
        print(f"{'═'*62}")

        for t in range(self.T - 1, -1, -1):
            print(f"\n{'─'*50}")
            print(f"  [t = {t}]  Generating {self.n_samples} Bellman targets ...")

            # Sample diverse states: W ~ LogNormal, p ~ Dirichlet
            W_samps = np.random.lognormal(mean=0.0, sigma=0.4, size=self.n_samples)
            p_samps = np.random.dirichlet(np.ones(self.n_assets), size=self.n_samples)

            X_list, y_list = [], []
            for i in range(self.n_samples):
                if (i + 1) % 100 == 0:
                    print(f"    state {i+1:4d}/{self.n_samples}")
                dp, val = self.optimize_delta_p(W_samps[i], p_samps[i], t)
                X_list.append(self._features(W_samps[i], p_samps[i]))
                y_list.append(val)

            X = np.array(X_list, dtype=np.float32)
            y = np.array(y_list, dtype=np.float32)

            # Standardise targets for stable NN training
            y_mu  = float(y.mean())
            y_sig = float(y.std()) + 1e-8
            y_nrm = (y - y_mu) / y_sig

            print(f"  [t = {t}]  Bellman target: "
                  f"mean={y_mu:.5f}  std={y_sig:.5f}  "
                  f"range=[{y.min():.4f}, {y.max():.4f}]")
            print(f"  [t = {t}]  Fitting ValueNet ...")

            X_t = torch.from_numpy(X).to(self.device)
            y_t = torch.from_numpy(y_nrm).to(self.device)

            net   = ValueNet(input_dim, hidden_dims=[128, 128, 64]).to(self.device)
            opt   = optim.Adam(net.parameters(), lr=self.lr, weight_decay=1e-5)
            sched = optim.lr_scheduler.StepLR(opt, step_size=150, gamma=0.5)
            loader = torch.utils.data.DataLoader(
                torch.utils.data.TensorDataset(X_t, y_t),
                batch_size=self.batch_size, shuffle=True,
            )

            for epoch in range(self.n_epochs):
                for xb, yb in loader:
                    opt.zero_grad()
                    nn.MSELoss()(net(xb), yb).backward()
                    opt.step()
                sched.step()
                if (epoch + 1) % 100 == 0:
                    with torch.no_grad():
                        loss = nn.MSELoss()(net(X_t), y_t).item()
                    print(f"    Epoch {epoch+1:4d}/{self.n_epochs}  MSE = {loss:.6f}")

            self.value_nets[t] = (net, y_mu, y_sig)

        print("\n✓ FVI training complete.\n")

    # ─────────────────────────────────────────────────────────────
    # Policy execution
    # ─────────────────────────────────────────────────────────────

    def execute_policy(
        self,
        W0: float = 1.0,
        p0: Optional[np.ndarray] = None,
        seed: int = 999,
    ) -> Dict:
        """
        Run greedy FVI policy for one simulated sample path.

        p0 : initial portfolio weights  [p_cash, p_1, …, p_n]
             Must satisfy Σ p0[k] = 1 and p0[k] ≥ 0.
             Default: equal weight across all n+1 assets.
        """
        assert self.value_nets, "Call train() before execute_policy()."
        np.random.seed(seed)

        if p0 is None:
            p0 = np.ones(self.n_assets) / self.n_assets
        assert np.isclose(p0.sum(), 1.0, atol=1e-5), "p0 must sum to 1."
        assert (p0 >= 0).all(),                       "p0[k] must be ≥ 0."

        labels = ['Cash'] + [f'Risky{k}' for k in range(1, self.n_risky + 1)]
        W, p   = float(W0), p0.copy()
        hist   = {'wealth': [W], 'weights': [p.copy()], 'actions': []}

        print(f"\n{'─'*55}")
        print(f"  Greedy policy execution  (W0 = {W0})")
        print(f"  Init: { {l: round(float(v),3) for l,v in zip(labels,p)} }")
        print(f"{'─'*55}")

        for t in range(self.T):
            dp, _     = self.optimize_delta_p(W, p, t)
            hist['actions'].append(dp.copy())
            risky_ret = self._sample_returns(1)[0]
            W_new, p_new = self.next_state(W, p, dp, risky_ret)

            if W_new is None:
                print(f"  [t={t}]  Bankrupt!")
                break

            print(f"  t={t}: W {W:.4f} → {W_new:.4f}   "
                  f"ret = {np.round(risky_ret,3)}   "
                  f"Δp = {np.round(dp,3)}")
            W, p = W_new, p_new
            hist['wealth'].append(W)
            hist['weights'].append(p.copy())

        print(f"\n  Final W    = {hist['wealth'][-1]:.4f}")
        print(f"  Final wts  = { {l: round(float(v),3) for l,v in zip(labels,hist['weights'][-1])} }")
        return hist

    # ─────────────────────────────────────────────────────────────
    # Analytic Merton benchmark  (Section 8.4)
    # ─────────────────────────────────────────────────────────────

    def analytic_merton(self, W: float = 1.0) -> np.ndarray:
        """
        Section 8.4 CARA unconstrained optimal allocation:
            p*_risky = (1/(α·W)) · Σ^{−1} · (a − r·1)

        The signal is excess return (a − r): higher excess → larger allocation.
        Negative weights clipped to 0 (no short-selling).
        Cash absorbs the remainder: p*_cash = 1 − Σ p*_risky.
        Returns full weight vector [p_cash, p_1, …, p_n].
        """
        excess  = self.a - self.r                              # (n_risky,)
        p_risky = np.linalg.solve(self.cov, excess) / (self.alpha * W)
        p_risky = np.maximum(p_risky, 0.0)                    # no shorting
        if p_risky.sum() > 1.0:
            p_risky /= p_risky.sum()                          # de-leverage
        p_cash = np.array([max(0.0, 1.0 - p_risky.sum())])
        return np.concatenate([p_cash, p_risky])

    # ─────────────────────────────────────────────────────────────
    # Baseline action functions
    # ─────────────────────────────────────────────────────────────

    def _fvi_action(self, W: float, p: np.ndarray, t: int,
                    n_mc_eval: int = 100) -> np.ndarray:
        dp, _ = self.optimize_delta_p(W, p, t, n_mc=n_mc_eval)
        return dp

    def _merton_action(self, W: float, p: np.ndarray, t: int) -> np.ndarray:
        """
        Move toward Merton target in one step (unconstrained — ignores delta_max).
        Since both target and p sum to 1, Σ Δp = 0 is automatically satisfied.
        """
        return self.analytic_merton(W) - p

    def _equalweight_action(self, W: float, p: np.ndarray, t: int) -> np.ndarray:
        """
        Move toward 1/(n+1) equal weight, clipped to ±delta_max.
        Re-enforces Σ Δp = 0 after clipping by subtracting the mean.
        """
        target = np.ones(self.n_assets) / self.n_assets
        dp     = np.clip(target - p, -self.delta_max, self.delta_max)
        dp    -= dp.mean()     # re-enforce Σ = 0 after clipping
        return dp

    # ─────────────────────────────────────────────────────────────
    # Monte Carlo strategy evaluation  (Common Random Numbers)
    # ─────────────────────────────────────────────────────────────

    def evaluate_strategy(
        self,
        strategy_fn: Callable,
        n_paths: int = 300,
        W0: float = 1.0,
        p0: Optional[np.ndarray] = None,
        seed: int = 0,
        label: str = '',
    ) -> Dict:
        """
        Evaluate strategy over n_paths Monte Carlo paths.

        Common Random Numbers (CRN): market return sequences are pre-generated
        with a fixed seed so all strategies face identical shocks.
        This eliminates market-noise variance from comparisons.
        """
        if p0 is None:
            p0 = np.ones(self.n_assets) / self.n_assets

        np.random.seed(seed)
        outer = self._sample_returns(n_paths * self.T).reshape(
            n_paths, self.T, self.n_risky
        )

        terminal_u, terminal_w = [], []
        paths = np.zeros((n_paths, self.T + 1))

        for i in range(n_paths):
            W, p = float(W0), p0.copy()
            paths[i, 0] = W
            bankrupt = False
            for t in range(self.T):
                dp = strategy_fn(W, p, t)
                W_new, p_new = self.next_state(W, p, dp, outer[i, t])
                if W_new is None:
                    bankrupt = True
                    break
                W, p = W_new, p_new
                paths[i, t + 1] = W
            terminal_u.append(-1e4 / self.alpha if bankrupt else self.utility(W))
            terminal_w.append(0.0               if bankrupt else W)

        tu, tw = np.array(terminal_u), np.array(terminal_w)
        alive  = tw > 0
        ew     = float(tw[alive].mean()) if alive.any() else 0.0
        sw     = float(tw[alive].std())  if alive.any() else 0.0

        print(f"  [{label:<22s}]  E[U] = {tu.mean():>10.5f}  "
              f"E[W_T] = {ew:>7.4f}  Std = {sw:>7.4f}  "
              f"Bankrupt = {int((~alive).sum())}/{n_paths}")

        return {
            'label': label,
            'E_utility':   float(tu.mean()),  'std_utility': float(tu.std()),
            'E_wealth':    ew,                 'std_wealth':  sw,
            'terminal_utilities': tu,          'terminal_wealths':  tw,
            'mean_path':   paths.mean(axis=0), 'paths_array': paths,
        }

    def compare_strategies(
        self,
        n_paths:  int   = 300,
        W0:       float = 1.0,
        seed:     int   = 0,
        n_mc_fvi: int   = 100,
    ) -> Dict[str, Dict]:
        """
        MC comparison: FVI (constrained) vs Merton (analytic) vs Equal Weight.
        All strategies face identical market shocks (CRN).
        """
        assert self.value_nets, "Call train() first!"

        print(f"\n{'═'*65}")
        print("  Monte Carlo Strategy Comparison")
        print(f"  n_paths = {n_paths}  |  seed = {seed}  |  n_mc_fvi = {n_mc_fvi}")
        print(f"{'═'*65}\n")

        labels   = ['Cash'] + [f'Risky{k}' for k in range(1, self.n_risky + 1)]
        p_merton = self.analytic_merton(W0)
        print(f"  Merton unconstrained target (W={W0}):")
        for lab, wt in zip(labels, p_merton):
            print(f"    {lab:>8s}: {wt:.4f}")

        strategies = {
            'FVI (constrained)':
                lambda W, p, t: self._fvi_action(W, p, t, n_mc_eval=n_mc_fvi),
            'Merton (projected)':
                lambda W, p, t: self._merton_action(W, p, t),
            'Equal Weight':
                lambda W, p, t: self._equalweight_action(W, p, t),
        }

        results = {}
        for name, fn in strategies.items():
            print(f"\n  ▶ {name} ...")
            results[name] = self.evaluate_strategy(
                fn, n_paths=n_paths, W0=W0, seed=seed, label=name
            )

        # Summary table
        print(f"\n{'─'*68}")
        print(f"  {'Strategy':<23} {'E[U(W_T)]':>12} "
              f"{'E[W_T]':>9} {'Std[W_T]':>10} {'Sharpe':>8}")
        print(f"{'─'*68}")
        for name, res in results.items():
            sharpe = res['E_wealth'] / (res['std_wealth'] + 1e-8)
            print(f"  {name:<23} {res['E_utility']:>12.5f} "
                  f"{res['E_wealth']:>9.4f} {res['std_wealth']:>10.4f} {sharpe:>8.3f}")
        print(f"{'─'*68}")

        # Paired t-tests (CRN enables valid pairing)
        fvi_u = results['FVI (constrained)']['terminal_utilities']
        n     = len(fvi_u)
        print()
        for name in ['Merton (projected)', 'Equal Weight']:
            diff   = fvi_u - results[name]['terminal_utilities']
            t_stat = diff.mean() / (diff.std() / np.sqrt(n) + 1e-10)
            print(f"  t-stat (FVI vs {name:<22}): {t_stat:>7.3f}  "
                  f"[{'FVI better' if t_stat > 0 else 'FVI worse'}]")

        return results

    # ─────────────────────────────────────────────────────────────
    # Visualisation
    # ─────────────────────────────────────────────────────────────

    def plot_results(self, hist: Dict, save_path: Optional[str] = None):
        labels = ['Cash'] + [f'Risky {k}' for k in range(1, self.n_risky + 1)]
        colors = ['#aaaaaa', 'steelblue', 'tomato', 'seagreen', 'orange'][:self.n_assets]

        fig, axes = plt.subplots(1, 3, figsize=(15, 4))

        # Wealth path
        axes[0].plot(hist['wealth'], 'b-o', lw=2, ms=7)
        axes[0].set_title('Wealth over Time', fontsize=12)
        axes[0].set_xlabel('Period $t$')
        axes[0].set_ylabel('$W_t$')
        axes[0].grid(alpha=0.3)

        # Portfolio weights
        wts = np.array(hist['weights'])
        for k, (lab, col) in enumerate(zip(labels, colors)):
            axes[1].plot(wts[:, k], '-o', color=col, label=lab, lw=1.8, ms=6)
        axes[1].set_title('Portfolio Weights', fontsize=12)
        axes[1].set_xlabel('Period $t$')
        axes[1].set_ylabel('$p_k$')
        axes[1].legend(fontsize=9)
        axes[1].grid(alpha=0.3)

        # Rebalancing actions
        if hist['actions']:
            acts = np.array(hist['actions'])
            nt   = len(acts)
            w    = 0.7 / self.n_assets
            x    = np.arange(nt)
            for k, (lab, col) in enumerate(zip(labels, colors)):
                axes[2].bar(x + (k - self.n_assets / 2 + 0.5) * w,
                            acts[:, k], w, color=col, label=lab, alpha=0.8)
            axes[2].axhline( self.delta_max, color='r', ls='--', lw=1.5,
                             label=f'±{self.delta_max:.0%}')
            axes[2].axhline(-self.delta_max, color='r', ls='--', lw=1.5)
            axes[2].axhline(0, color='k', lw=0.5, alpha=0.3)
            axes[2].set_title('Rebalancing $\\Delta p$', fontsize=12)
            axes[2].set_xlabel('Period $t$')
            axes[2].set_ylabel('$\\Delta p_k$')
            axes[2].legend(fontsize=8)
            axes[2].grid(alpha=0.3)

        plt.suptitle('FVI Portfolio Optimization — Single Simulated Path', fontsize=13)
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path, dpi=120, bbox_inches='tight')
            print(f"  Saved: {save_path}")
        plt.show()

    def plot_comparison(
        self,
        results: Dict[str, Dict],
        save_path: Optional[str] = None,
    ):
        names   = list(results.keys())
        colors  = ['steelblue', 'tomato', 'seagreen']
        n_paths = len(results[names[0]]['terminal_utilities'])

        fig, axes = plt.subplots(1, 3, figsize=(15, 5))

        # (a) E[U(W_T)] bar chart + 95% CI error bars
        eu  = [results[n]['E_utility']                    for n in names]
        seu = [results[n]['std_utility'] / np.sqrt(n_paths) for n in names]
        axes[0].bar(names, eu, color=colors, alpha=0.8, edgecolor='k', lw=0.8)
        axes[0].errorbar(names, eu, yerr=[1.96 * s for s in seu],
                         fmt='none', color='k', capsize=6, lw=1.5)
        axes[0].set_title('$\\mathbb{E}[U(W_T)]$ with 95% CI', fontsize=11)
        axes[0].set_ylabel('$\\mathbb{E}[U(W_T)]$')
        axes[0].tick_params(axis='x', rotation=15)
        axes[0].grid(axis='y', alpha=0.3)

        # (b) Terminal wealth box plot
        bp = axes[1].boxplot(
            [results[n]['terminal_wealths'] for n in names],
            patch_artist=True,
            medianprops={'color': 'k', 'lw': 2},
            flierprops={'marker': 'o', 'ms': 3, 'alpha': 0.4},
        )
        for patch, c in zip(bp['boxes'], colors):
            patch.set_facecolor(c)
            patch.set_alpha(0.7)
        axes[1].set_xticklabels(names, rotation=15, ha='right')
        axes[1].set_title('Terminal Wealth $W_T$', fontsize=11)
        axes[1].set_ylabel('$W_T$')
        axes[1].grid(axis='y', alpha=0.3)

        # (c) Mean wealth paths + 10–90th percentile bands
        t_ax = np.arange(self.T + 1)
        for name, c in zip(names, colors):
            paths = results[name]['paths_array']
            axes[2].plot(t_ax, results[name]['mean_path'], '-o',
                         color=c, label=name, lw=2, ms=5)
            axes[2].fill_between(
                t_ax,
                np.percentile(paths, 10, axis=0),
                np.percentile(paths, 90, axis=0),
                color=c, alpha=0.13,
            )
        axes[2].set_title('Mean Wealth Path (shaded: 10–90 pct)', fontsize=11)
        axes[2].set_xlabel('Period $t$')
        axes[2].set_ylabel('$\\mathbb{E}[W_t]$')
        axes[2].legend(fontsize=9)
        axes[2].grid(alpha=0.3)

        plt.suptitle(f'Monte Carlo Strategy Comparison  (n_paths = {n_paths})',
                     fontsize=13)
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path, dpi=120, bbox_inches='tight')
            print(f"  Saved: {save_path}")
        plt.show()


# ════════════════════════════════════════════════════════════════════
#  Demo — Assignment 1
#  Shows the solver works for any reasonable (n, T, r, a, s, p0)
# ════════════════════════════════════════════════════════════════════

if __name__ == '__main__':

    # ── Market parameters (per assignment: s[k] = VARIANCE) ──────
    a    = np.array([0.08, 0.12, 0.10])    # mean per-period returns
    s    = np.array([0.04, 0.09, 0.06])    # variances  (std ≈ 0.20, 0.30, 0.24)
    corr = np.array([[1.00, 0.30, 0.20],
                     [0.30, 1.00, 0.35],
                     [0.20, 0.35, 1.00]])

    # ── 1. Create solver ──────────────────────────────────────────
    #   n_risky=3  →  assets {cash, risky1, risky2, risky3}
    #   Constraints: n > 2 ✓   n < 5 ✓   T < 10 ✓
    fvi = FVI(
        n_risky   = 3,
        T         = 5,
        alpha     = 2.0,
        delta_max = 0.10,
        r         = 0.03,
        a=a, s=s, corr=corr,
        seed      = 42,
    )

    # ── 2. Train ──────────────────────────────────────────────────
    fvi.train(n_samples=500, n_mc=300, n_epochs=400, batch_size=64, lr=1e-3)

    # ── 3. Execute one path with user-specified initial portfolio ─
    #   p0 = [cash, risky1, risky2, risky3],   must sum to 1
    p0 = np.array([0.25, 0.40, 0.25, 0.10])
    history = fvi.execute_policy(W0=1.0, p0=p0)
    fvi.plot_results(history, save_path='fvi_path.png')

    # ── 4. Analytic Merton benchmark (Section 8.4) ────────────────
    p_merton = fvi.analytic_merton(W=1.0)
    print(f"\n{'─'*55}")
    print(f"  Merton unconstrained (W=1) : {np.round(p_merton, 4)}")
    print(f"  FVI final weights          : {np.round(history['weights'][-1], 4)}")
    print(f"{'─'*55}")

    # ── 5. Monte Carlo strategy comparison ───────────────────────
    #   Quick test  : n_paths=50,  n_mc_fvi=50   (~1–2 min)
    #   Full eval   : n_paths=300, n_mc_fvi=100  (~10–20 min)
    comparison = fvi.compare_strategies(n_paths=300, W0=1.0, seed=0, n_mc_fvi=100)
    fvi.plot_comparison(comparison, save_path='fvi_comparison.png')

    # # ── 6. Demonstrate generality: sweep over (n, T, r) configs ──
    # print("\n" + "═" * 62)
    # print("  Generality check: different (n_risky, T, r) configurations")
    # print("═" * 62)

    # configs = [
    #     dict(n_risky=3, T=3, r=0.02),   # short horizon
    #     dict(n_risky=4, T=5, r=0.04),   # 4 risky assets
    #     dict(n_risky=3, T=8, r=0.01),   # long horizon, low r
    #     dict(n_risky=4, T=9, r=0.05),   # near max T and n
    # ]

    # for cfg in configs:
    #     n, T, r = cfg['n_risky'], cfg['T'], cfg['r']
    #     _a   = np.array([0.08, 0.12, 0.10, 0.09])[:n]
    #     _s   = np.array([0.04, 0.09, 0.06, 0.05])[:n]
    #     _fvi = FVI(n_risky=n, T=T, alpha=2.0, delta_max=0.10,
    #                r=r, a=_a, s=_s, seed=0)
    #     _fvi.train(n_samples=300, n_mc=200, n_epochs=200)
    #     _h = _fvi.execute_policy(W0=1.0, seed=1)
    #     print(f"  n={n}, T={T}, r={r:.0%}  →  final W = {_h['wealth'][-1]:.4f}")
```

```python
# ═══════════════════════════════════════════════════════════════
# FVI 验证实验
# ═══════════════════════════════════════════════════════════════
# 
# **核心原则：从有已知答案的简单情形开始，逐步增加复杂度。**
# 
# | 层级 | 实验 | 目标 |
# |------|------|------|
# | 第一层：算法正确性 | 实验1: 单资产无约束 | FVI ≈ Merton 解析解？ |
# |  | 实验2: γ 扫描 | 持仓是否单调递减？ |
# |  | 实验3: 放松约束 | 约束是否导致三策略无差异？ |
# | 第二层：策略质量 | 实验4: 极端资产差异 | FVI 能否超配好资产？ |
# |  | 实验5: 延长 T | 复利能否放大差异？ |
# | 第三层：统计功效 | 实验6: 增大路径数 | 策略差距是否稳定可检出？ |
```

```python
# ══════════════════════════════════════════════════════════════════
# 实验 1: 单资产 + 无约束 — 黄金标准测试
# ══════════════════════════════════════════════════════════════════
# 目的: 1 个风险资产 + CARA 效用 + 无调仓约束 → 有 Merton 精确解
#       p* = (1 / (α·W)) · μ / σ²
#       如果 FVI ≈ p*，说明算法正确

print("=" * 65)
print("  实验 1: 单资产 + 无约束 (Merton 精确解对比)")
print("=" * 65)

fvi_exp1 = FVI(
    n_assets=1,
    T=5,
    alpha=1.0,
    delta_max=10.0,         # 放到很大 → 等效无约束
    mu=np.array([0.10]),
    sigma=np.array([0.20]),
    corr=np.array([[1.0]]),
    dt=1.0 / 12,
    seed=42
)

fvi_exp1.train(n_samples=300, n_mc=80, n_epochs=300, batch_size=64, lr=1e-3)

# 在不同财富水平验证
print(f"\n{'W':>6} | {'Merton p*':>12} | {'FVI p':>12} | {'|error|':>10}")
print("─" * 50)

for W_test in [0.5, 1.0, 1.5, 2.0, 3.0]:
    p_init = np.array([0.5])
    dp_fvi, _ = fvi_exp1.optimize_delta_p(W_test, p_init, t=0, n_mc=200)
    p_fvi = p_init + dp_fvi

    p_merton = fvi_exp1.analytic_merton(W=W_test)
    error = abs(p_fvi[0] - p_merton[0])
    print(f"{W_test:>6.1f} | {p_merton[0]:>12.4f} | {p_fvi[0]:>12.4f} | {error:>10.4f}")

print("\n✓ 如果 |error| 很小 → FVI 算法正确")
print("✗ 如果 |error| 很大 → 算法本身有问题，不要继续后续实验")
```

```python
# ══════════════════════════════════════════════════════════════════
# 实验 2: 风险厌恶系数 γ(α) 扫描
# ══════════════════════════════════════════════════════════════════
# 目的: α 越大 → 越厌恶风险 → 持仓应该单调递减
#       验证 FVI 对参数变化的响应是否符合理论方向

print("=" * 65)
print("  实验 2: 风险厌恶系数 α 扫描")
print("=" * 65)

alphas = [0.5, 1.0, 3.0, 10.0]
exp2_results = {}

for alpha_val in alphas:
    print(f"\n── α = {alpha_val} ──")
    fvi_tmp = FVI(
        n_assets=1,
        T=5,
        alpha=alpha_val,
        delta_max=10.0,         # 无约束
        mu=np.array([0.10]),
        sigma=np.array([0.20]),
        corr=np.array([[1.0]]),
        dt=1.0 / 12,
        seed=42
    )
    fvi_tmp.train(n_samples=200, n_mc=60, n_epochs=200, batch_size=64, lr=1e-3)

    p_init = np.array([0.5])
    dp, _ = fvi_tmp.optimize_delta_p(1.0, p_init, t=0, n_mc=200)
    p_fvi = (p_init + dp)[0]
    p_merton = fvi_tmp.analytic_merton(W=1.0)[0]
    exp2_results[alpha_val] = {'fvi': p_fvi, 'merton': p_merton}

# 汇总
print(f"\n{'α':>6} | {'Merton p*':>12} | {'FVI p':>12} | {'单调?':>6}")
print("─" * 50)
prev_p = None
for alpha_val in alphas:
    r = exp2_results[alpha_val]
    monotone = ""
    if prev_p is not None:
        monotone = "✓" if r['fvi'] < prev_p else "✗"
    prev_p = r['fvi']
    print(f"{alpha_val:>6.1f} | {r['merton']:>12.4f} | {r['fvi']:>12.4f} | {monotone:>6}")

print("\n✓ FVI 持仓随 α 单调递减 → 参数响应正确")
print("✗ 非单调 → 训练不够充分或网络拟合问题")
```

```python
# ══════════════════════════════════════════════════════════════════
# 实验 3: 调仓约束逐步放松
# ══════════════════════════════════════════════════════════════════
# 目的: 约束从紧到松，FVI 应该从等权走向 Merton
#   δ_max = 1%   → FVI ≈ Equal Weight（几乎被锁死）
#   δ_max = 10%  → FVI 介于中间
#   δ_max = 100% → FVI ≈ Merton

print("=" * 65)
print("  实验 3: 调仓约束逐步放松 → 策略差异变化")
print("=" * 65)

delta_maxes = [0.01, 0.05, 0.10, 0.50, 1.00, 10.0]
exp3_results = {}

mu_3 = np.array([0.10])
sigma_3 = np.array([0.20])
corr_3 = np.array([[1.0]])

for dm in delta_maxes:
    print(f"\n── δ_max = {dm} ──")
    fvi_tmp = FVI(
        n_assets=1, T=5, alpha=1.0,
        delta_max=dm,
        mu=mu_3, sigma=sigma_3, corr=corr_3,
        dt=1.0 / 12, seed=42
    )
    fvi_tmp.train(n_samples=200, n_mc=60, n_epochs=200, batch_size=64, lr=1e-3)

    # 比较策略
    comparison_tmp = fvi_tmp.compare_strategies(
        n_paths=200, W0=1.0, seed=0, n_mc_fvi=30
    )
    exp3_results[dm] = {
        name: res['E_utility'] for name, res in comparison_tmp.items()
    }

# 汇总
print(f"\n{'δ_max':>8} | {'E[U] FVI':>12} | {'E[U] Merton':>13} | {'E[U] EqW':>11} | {'FVI-EqW':>9}")
print("─" * 65)
for dm in delta_maxes:
    r = exp3_results[dm]
    fvi_u = r.get('FVI (constrained)', 0)
    mer_u = r.get('Merton (projected)', 0)
    eqw_u = r.get('Equal Weight', 0)
    diff = fvi_u - eqw_u
    print(f"{dm:>8.2f} | {fvi_u:>12.6f} | {mer_u:>13.6f} | {eqw_u:>11.6f} | {diff:>+9.6f}")

print("\n✓ δ_max 增大 → FVI 与 Merton 差距缩小，与 EqW 差距增大")
print("  说明之前三策略无差异是约束造成的，模型本身正确")
```

```python
# ══════════════════════════════════════════════════════════════════
# 实验 4: 制造明显的最优持仓（极端资产差异）
# ══════════════════════════════════════════════════════════════════
# 目的: 资产 1 回报极高，资产 2 回报极低 → 最优应大量超配资产 1
#       如果 FVI 仍然接近等权 → 值函数拟合有问题

print("=" * 65)
print("  实验 4: 极端资产差异 → FVI 能否正确超配好资产？")
print("=" * 65)

fvi_exp4 = FVI(
    n_assets=2,
    T=5,
    alpha=1.0,
    delta_max=10.0,         # 无约束，让策略自由发挥
    mu=np.array([0.15, 0.02]),    # 资产 1 远优于 资产 2
    sigma=np.array([0.20, 0.20]), # 波动率相同
    corr=np.array([[1.0, 0.0],
                   [0.0, 1.0]]),  # 不相关
    dt=1.0 / 12,
    seed=42
)

fvi_exp4.train(n_samples=300, n_mc=80, n_epochs=300, batch_size=64, lr=1e-3)

# 对比各策略
comparison_exp4 = fvi_exp4.compare_strategies(
    n_paths=300, W0=1.0, seed=0, n_mc_fvi=30
)

# 查看 FVI 在 t=0 对仓位的选择
p_init = np.array([0.5, 0.5])  # 从等权开始
dp, _ = fvi_exp4.optimize_delta_p(1.0, p_init, t=0, n_mc=200)
p_fvi = p_init + dp
p_merton = fvi_exp4.analytic_merton(W=1.0)

print(f"\nMerton 解析最优: {np.round(p_merton, 4)}")
print(f"FVI 最优仓位:    {np.round(p_fvi, 4)}")
print(f"等权基准:        [0.5000, 0.5000]")
print(f"\n✓ FVI 应该明显超配资产 1 (p₁ >> p₂)")
print(f"  资产 1 权重 = {p_fvi[0]:.4f},  资产 2 权重 = {p_fvi[1]:.4f}")

fvi_exp4.plot_comparison(comparison_exp4, save_path='exp4_extreme_assets.png')
```

```python
# ══════════════════════════════════════════════════════════════════
# 实验 5: 延长时间跨度 T
# ══════════════════════════════════════════════════════════════════
# 目的: T 越长 → 复利放大策略差异 → 蒙特卡洛更容易区分
#       T=1 几乎无差异,  T=30 差异应该很显著

print("=" * 65)
print("  实验 5: 延长时间跨度 T → 复利放大策略差异")
print("=" * 65)

T_values = [1, 5, 10, 20]
exp5_results = {}

for T_val in T_values:
    print(f"\n{'═' * 60}")
    print(f"  T = {T_val}")
    print(f"{'═' * 60}")

    fvi_tmp = FVI(
        n_assets=2,
        T=T_val,
        alpha=1.0,
        delta_max=0.10,
        mu=np.array([0.10, 0.04]),
        sigma=np.array([0.20, 0.15]),
        corr=np.array([[1.0, 0.3], [0.3, 1.0]]),
        dt=1.0 / 12,
        seed=42
    )
    # T 大时减少样本加速（Colab 上可适当调大）
    ns = min(200, max(100, 300 // T_val))
    fvi_tmp.train(n_samples=ns, n_mc=60, n_epochs=200, batch_size=64, lr=1e-3)

    comp = fvi_tmp.compare_strategies(n_paths=200, W0=1.0, seed=0, n_mc_fvi=30)
    exp5_results[T_val] = {
        name: {'E_utility': res['E_utility'], 'E_wealth': res['E_wealth']}
        for name, res in comp.items()
    }

# 汇总
print(f"\n{'T':>4} | {'E[U] FVI':>12} | {'E[U] Merton':>13} | {'E[U] EqW':>11} | {'FVI−EqW':>10}")
print("─" * 60)
for T_val in T_values:
    r = exp5_results[T_val]
    fvi_u = r['FVI (constrained)']['E_utility']
    mer_u = r['Merton (projected)']['E_utility']
    eqw_u = r['Equal Weight']['E_utility']
    print(f"{T_val:>4} | {fvi_u:>12.6f} | {mer_u:>13.6f} | {eqw_u:>11.6f} | {fvi_u - eqw_u:>+10.6f}")

print("\n✓ T 增大 → FVI 与 EqW 的差距应该逐渐增大")
```

```python
# ══════════════════════════════════════════════════════════════════
# 实验 6: 增大蒙特卡洛路径数 n_paths
# ══════════════════════════════════════════════════════════════════
# 目的: 检验统计功效——置信区间是否随路径数缩小、策略差距是否稳定
#       n_paths: 50 → 200 → 1000 → 5000

print("=" * 65)
print("  实验 6: 增大 n_paths → 检验统计功效")
print("=" * 65)

# 复用实验 4 的模型（极端差异，差距明显）
# 如果 fvi_exp4 不在内存里则重新训练
try:
    _ = fvi_exp4.value_nets[0]
    print("复用实验 4 的模型")
except:
    print("重新训练实验 4 模型 ...")
    fvi_exp4 = FVI(
        n_assets=2, T=5, alpha=1.0, delta_max=10.0,
        mu=np.array([0.15, 0.02]),
        sigma=np.array([0.20, 0.20]),
        corr=np.array([[1.0, 0.0], [0.0, 1.0]]),
        dt=1.0 / 12, seed=42
    )
    fvi_exp4.train(n_samples=300, n_mc=80, n_epochs=300, batch_size=64, lr=1e-3)

path_counts = [50, 200, 1000, 5000]
exp6_results = {}

for np_val in path_counts:
    print(f"\n── n_paths = {np_val} ──")
    comp = fvi_exp4.compare_strategies(
        n_paths=np_val, W0=1.0, seed=0, n_mc_fvi=30
    )
    n = np_val
    exp6_results[np_val] = {}
    for name, res in comp.items():
        se = res['std_utility'] / np.sqrt(n)
        exp6_results[np_val][name] = {
            'E_utility': res['E_utility'],
            'SE': se,
            'CI_95': 1.96 * se
        }

# 汇总: 关注 FVI 的置信区间和与 EqW 的差距
print(f"\n{'n_paths':>8} | {'E[U] FVI':>12} | {'95%CI FVI':>11} | {'FVI−EqW':>10} | {'显著?':>5}")
print("─" * 60)
for np_val in path_counts:
    r = exp6_results[np_val]
    fvi_u = r['FVI (constrained)']['E_utility']
    fvi_ci = r['FVI (constrained)']['CI_95']
    eqw_u = r['Equal Weight']['E_utility']
    eqw_ci = r['Equal Weight']['CI_95']
    diff = fvi_u - eqw_u
    # 粗略判断: 差距 > 两个CI之和 → 显著
    sig = "✓" if abs(diff) > (fvi_ci + eqw_ci) else "?"
    print(f"{np_val:>8} | {fvi_u:>12.6f} | {fvi_ci:>11.6f} | {diff:>+10.6f} | {sig:>5}")

print("\n✓ n_paths 增大 → CI 缩小，差距稳定 → 策略差异真实存在")
print("? CI 仍然覆盖 0 → 需要更多路径或更大的设置差异")
```
